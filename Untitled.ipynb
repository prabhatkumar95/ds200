```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import csv
```

```python
data_path="data.csv"
```

```python
data = pd.read_csv(data_path)
```

```python
data.boxplot()
plt.xlabel("Compound")
plt.ylabel("Value (ppm)")
plt.title("BLR Air Q Statistics")
plt.savefig("plots/box_plot.png")
```

```python
keys = data.keys()
print(keys)
```

```python
data.shape
```

```python
for i in range(1,len(keys)):
    for j in range(i+1,len(keys)):
        data.plot.scatter(x = i,y=j)
        plt.xlabel(str(keys[i]))
        plt.ylabel(str(keys[j]))
        plt.title ("BLR Air Q Statistics : "+str(keys[j])+" - "+str(keys[i]))
        plt.savefig("plots/scatter"+str(keys[i])+"_"+str(keys[j])+".png")
```
